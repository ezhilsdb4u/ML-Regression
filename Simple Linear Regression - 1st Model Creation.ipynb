```python
#Pandas is a library that will be used when we handle datasets in tables or files
#We can call all functions inside pandas library if we import it
#pd is just a alias for pandas
import pandas as pd
```

```python
#Read_csv is a function in pandas 
#dataset is a varible. All data from the file is now stored in dataset.
dataset = pd.read_csv("Salary_Data.csv")
```

```python
#we are calling this variable to check the data. We can either call it as is or we can even print(dataset) to check the output.
#as a defult it is coming with row number starting at 0
print(dataset)
```

```python
#Consider Independent as Input variable (Which is X value in Y=MX+C)
#Assign "YearsExperience" to independent variable
independent = dataset[["YearsExperience"]]
```

```python
#everytime first value gets '0' rownumber
print(independent)
```

```python
#Consider dependent as Output variable (Which is Y value in Y=MX+C)
#Assign "Salary" to dependent variable
dependent = dataset[["Salary"]]  
```

```python
print(dependent)
```

```python
#sklearn is library which support all functions for ML.
#Model_selection is a sub folder in that library.
#train_test_split is a function within the library which is used to split the dataset into train set and test set
#Here we are importing this function under Sklearn so that we can save memory. we are not importing all functions.
from sklearn.model_selection import train_test_split
```

```python
#We need to give 4 parameters to this function
#based on then test_size it splits the data as below. 
print(train_test_split(independent, dependent,test_size=0.50, random_state=0))
```

```python
#Assign variables to each sections -> always 1st-input train, 2nd - input test, 3rd - output train, 4th - output test 
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent,test_size=0.50, random_state=0)
```

```python
Y_test
```

```python
#sklearn is library which support all functions for ML.
#linear_model is a sub folder in that library.
#LinearRegression is a function within the library which is used to build the model for liner regression
#Here we are importing this function under Sklearn so that we can save memory. we are not importing all functions.
from sklearn.linear_model import LinearRegression
#regressor is a variable and assigne LinearRegression () class to this
regressor=LinearRegression()
#fit is a function in LinearRegression
regressor.fit(X_train,Y_train)
#With this step model is created
```

```python
Slope=regressor.coef_
#This is to identify the Slope
```

```python
Slope
```

```python
bias=regressor.intercept_
```

```python
bias
#This is to ideitify the starting point which is bias
```

```python
Y_Pred=regressor.predict(X_test)
#This is to get the precited value  for the given test data
```

```python
Y_Pred
```

```python
print("Actual:", Y_test)
print("Predicted:", Y_Pred)
```

```python
#This to do the validation how good our prediction is. IF close to 1 it is good
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_Pred)
```

```python
r_score
```

```python
#Pickle is the library to save the model we created.
import pickle
filename="Finalized_linear_model.sav"
```

```python
#in this step we are dumping the created model.. we are storing the model with wb.. write binary
pickle.dump(regressor, open(filename,'wb'))
```

```python
#IF we want to load the model to execute use rb (read binary) that is read binary
load_model=pickle.load(open('Finalized_linear_model.sav','rb'))
result=load_model.predict([[15]])
```
